```python
#QUERY 1 - Dias com neblina
q = """
    SELECT fog, max(maxtempi)
    FROM weather_data
    GROUP BY fog
    """
foggy_days = psql.sqldf(q.lower(), locals())
foggy_days.index = ['Clear', 'Foggy']
foggy_days
```

```python
# QUERY 2 - Dias de chuva
weather_data = pd.read_csv('weather-underground.csv')
q = """
    SELECT count(rain)
    FROM weather_data
    WHERE rain=1
    """
rainy_days = psql.sqldf(q.lower(), locals())
rainy_days.index = ['Rainy Days']
rainy_days
```

```python
# QUERY 3 - Temperatura média aos sábados e domingos
q = """
    SELECT avg(meantempi)
    FROM weather_data
    WHERE cast(strftime('%w', date) as integer) = 0 or cast(strftime('%w', date) as integer) = 6;
    """
mean_temp_weekends = psql.sqldf(q.lower(), locals())
mean_temp_weekends.index = ['Average Temperature']
mean_temp_weekends
```

```python
#QUERY 4 - Temperatura mínima média em dias de chuva com temp mínimi 55°
q = """
    SELECT avg(mintempi)
    from weather_data
    WHERE rain = 1 and mintempi>55
    """
avg_min_temp_rainy = psql.sqldf(q.lower(), locals())
avg_min_temp_rainy.index = ['Temperatura Mínima']
avg_min_temp_rainy
```
